# Exploratory Data Analysis

# Salary Analysis

# Experience Level

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='experience_level', y='salary_in_usd')

plt.title('Salary Distribution by Experience Level', fontsize=14)
plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Salary in USD', fontsize=12)

plt.show()
```

# Employement Type

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='employment_type', y='salary_in_usd')

plt.title('Salary Distribution by Employment Type', fontsize=14)
plt.xlabel('Employment Type', fontsize=12)
plt.ylabel('Salary in USD', fontsize=12)

plt.show()
```

# Salary Distribution

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='job_title', y='salary_in_usd')

plt.xticks(rotation=90)

plt.title('Salary Distribution by Job Title', fontsize=14)
plt.xlabel('Job Title', fontsize=12)
plt.ylabel('Salary in USD', fontsize=12)

plt.show()
```

# Salary Distribution in USD

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['salary_in_usd'], kde=True, bins=30)

plt.title('Salary Distribution in USD', fontsize=14)
plt.xlabel('Salary in USD', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.show()
```

# Local Salary VS USD Salary Comparison

```python
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')

plt.figure(figsize=(10, 6))

sns.scatterplot(data=df, x='salary', y='salary_in_usd', hue='salary_currency')

plt.title('Comparison of Local Salary vs Salary in USD', fontsize=14)
plt.xlabel('Local Salary', fontsize=12)
plt.ylabel('Salary in USD', fontsize=12)

plt.show()
```

# Impact of Remote Ratio on Salaries

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='remote_ratio', y='salary_in_usd')

plt.title('Impact of Remote Ratio on Salaries', fontsize=14)
plt.xlabel('Remote Work Ratio (%)', fontsize=12)
plt.ylabel('Salary in USD', fontsize=12)

plt.show()
```

# Geographical Analysis

```python
plt.figure(figsize=(24, 10))
plt.subplot(1, 2, 1)
sns.barplot(data=df, x='company_location', y='salary_in_usd')
plt.xticks(rotation=45, ha='right', fontsize=8)  # Rotate and reduce x-axis label font size
plt.title('Salary Distribution by Company Location', fontsize=14)
plt.xlabel('Company Location', fontsize=12)
plt.ylabel('Salary in USD', fontsize=12)
plt.subplot(1, 2, 2)
sns.barplot(data=df, x='employee_residence', y='salary_in_usd')
plt.xticks(rotation=45, ha='right', fontsize=8)  # Rotate and reduce x-axis label font size
plt.title('Salary Distribution by Employee Residence', fontsize=14)
plt.xlabel('Employee Residence', fontsize=12)
plt.ylabel('Salary in USD', fontsize=12)
plt.tight_layout()
plt.show()
```

# Company Size vs Salary:

```python
print(df.columns)
salary_by_company_size = df.groupby('company_size')['salary'].agg(['mean', 'median', 'min', 'max', 'count']).reset_index()
print(salary_by_company_size)

plt.figure(figsize=(10, 6))
sns.barplot(x='company_size', y='salary', data=df, palette='Set2')
plt.title('Salary Distribution Across Different Company Sizes')
plt.xlabel('Company Size')
plt.ylabel('Salary')
plt.show()
```

# Percentage of Jobs

```python
plt.figure(figsize=(5,5))
d=(df["job_title"].value_counts (normalize=True)*100).head()
keys=df["job_title"].value_counts().head().index
colorz = ['#FF6347, #4682B4, #800000, #008000, #FFFF00']
exploda = (0.02, 0.02, 0.02, 0.4, 0.02)
plt.pie(d, labels=keys, autopct='%1.1f%%', explode=exploda, startangle=45, textprops={'fontsize': 18})
plt.savefig("audiencePie.png")
```

# World Cloud of Company Location

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

all_reviews = ' '.join(df['company_location'].dropna())

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_reviews)

plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide the axes
plt.title('COMPANY LOCATION')
plt.show()
```
