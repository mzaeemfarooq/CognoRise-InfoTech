# Engagement

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['Total_Accepted_Campaigns'] = df[['AcceptedCmp1', 
                                     'AcceptedCmp2', 'AcceptedCmp3', 
                                     'AcceptedCmp4', 'AcceptedCmp5']].sum(axis=1)

engagement_columns = ['Recency', 'NumWebVisitsMonth', 'Total_Accepted_Campaigns']

df[engagement_columns] = scaler.fit_transform(df[engagement_columns])

df['Engagement_Score'] = df['NumWebVisitsMonth'] * 0.4 + df['Total_Accepted_Campaigns'] * 0.4 - df['Recency'] * 0.2

df[['Engagement_Score']].head()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(df['Engagement_Score'], bins=20, kde=True, color='lightblue')

plt.title('Distribution of Customer Engagement Score', fontsize=16)
plt.xlabel('Engagement Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.show()
```

# Campaign Effectiveness Hypothesis Testing:

```python
from scipy import stats
median_income = df['Income'].median()
df['Income_Category'] = ['High' if income >= median_income else 'Low' for income in df['Income']]
df['Total_Accepted_Campaigns'] = df[['AcceptedCmp1', 'AcceptedCmp2', '
                                     'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']].sum(axis=1)
high_income_accepted = df[df['Income_Category'] == 'High']['Total_Accepted_Campaigns']
low_income_accepted = df[df['Income_Category'] == 'Low']['Total_Accepted_Campaigns']

# Performing an independent t-test to compare the two groups
t_stat, p_value = stats.ttest_ind(high_income_accepted, low_income_accepted)

# Output the t-statistic and p-value
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpret the result
if p_value < 0.05:
    print("Reject the null hypothesis: Higher-income customers are more likely to accept campaigns.")
else:
    print("Fail to reject the null hypothesis: No significant difference in campaign acceptance between income groups.")
```
