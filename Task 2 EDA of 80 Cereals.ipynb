```python
from wordcloud import WordCloud
```

```python
# Combine all the text from the 'name' column into one string
text = " ".join(name for name in df['name'].astype(str))

# Generate the Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the wordcloud
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off') #Turning of axis
plt.title('Word Cloud of Names', fontsize=20)
plt.show()
```

# Pie Chart

```python
plt.figure(figsize=(5,5))

d = df['name'].value_counts(normalize=True) * 100 
d = d.head()

keys = df['name'].value_counts().head().index
colorz= ['#B5DF00', '#AD1FFF', '#FFC93F', '#5FB1FF', '#BF1B00']

exploda= (0.02, 0.02, 0.02, 0.4, 0.02)

# Creating a pie chart
plt.pie(d, labels=keys, autopct='%1.1f%%', explode=exploda, startangle=45,
       textprops={'fontsize': 18}, colors=colorz)
plt.show()
```
